```python
import numpy as np
import matplotlib.pyplot as plt

h=20
v=10
g=9.81

T = np.sqrt(2*h/g)
t=np.linspace(0,T,num=100)
x=v*t
y=h-(g*t**2)/2

plt.figure(figsize=(10,6))
plt.plot(x,y,color="#8E44AD",linestyle="-.")
plt.grid(which="both")
plt.minorticks_on()
plt.title("H balandlikdan otilgan jismning harakat grafigi")
plt.xlabel("S, (Uzoqlik)")
plt.ylabel("H, (Balandlik)")
#plt.legend("chiziq", loc=3)
plt.legend(title='Funksiya grafigi', fontsize=10, frameon=True, facecolor='lightgray', edgecolor='black', shadow=True, loc=0)

plt.show()
```
