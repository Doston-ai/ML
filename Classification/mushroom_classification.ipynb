### # Ustunlarning tarqalishini vizualizatsiya qilish

```python
fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(20, 25))
axes = axes.flatten()

for i, col in enumerate(df.columns):
    sns.countplot(x=col, data=df, ax=axes[i], palette='viridis')
    axes[i].set_title(f'{col} ustunining tarqalishi')

# Foydalanilmagan subplotslarni o'chirish
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## MLga tayyorgarlik

```python
X = mushroom.drop("class", axis=1)
y = mushroom["class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# ML


## k-NN algoritmi

k-NN algoritmini boshlashdan oldin malumotlarimiz uchun eng yaxshi k ni topib olishimiz kerak bo'ladi

```python
f1 = []
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_predict = knn.predict(X_test)
  f1.append(f1_score(y_test, y_predict))

plt.figure(figsize=(12,6))
plt.plot(range(1,25), f1)
plt.xticks(range(1,25))
plt.grid(which="both")
plt.show()
```

` for ` yordamida ` k` ni qiymatini topganimizda eng yaxshi natija ` 1` chiqdi.

Keling endi ` sklearn` yordamida  ` k` ni eng yaxshi qiymatini topib olamiz

```python
param_grid = {"n_neighbors":np.arange(1,25)}
knn_gscv = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
knn_gscv.fit(X,y)
```

```python
knn_gscv.cv_results_
```

```python
knn_gscv.cv_results_["rank_test_score"]
```

```python
knn_gscv.best_params_
```

```python
plt.figure(figsize=(12,6))
plt.plot(param_grid['n_neighbors'], knn_gscv.cv_results_["rank_test_score"])
plt.xticks(param_grid['n_neighbors'])
plt.xlabel("k")
plt.ylabel("Xatolik reytingi")
plt.grid(linestyle="dotted")
plt.show()
```

Ikkala natijamiz ham bir xil ` 1` chiqdi.Endi k o'rniga qo'yib modelni baholaymiz

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_predict = knn.predict(X_test)
```

### Jaccard Index

```python
jaccar_index = jaccard_score(y_test, y_predict)
jaccar_index
```

### Confusion Metrics

```python
if len(y_test.shape) > 1:
    y_test = y_test.ravel()

if len(y_predict.shape) > 1:
    y_predict = y_predict.ravel()

cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
plt.show()
# Roc Curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_predict)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```

```python
clfn_report = classification_report(y_test, y_predict)
print(clfn_report)
```

```python
print("Precision:", precision_score(y_test, y_predict))
print("Recall:", recall_score(y_test, y_predict))
print("f1-score:", f1_score(y_test, y_predict))
```
